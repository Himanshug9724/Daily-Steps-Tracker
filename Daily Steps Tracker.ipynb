```python
import pandas as pd
import matplotlib.pyplot as plt

# CSV file se data read karna
df = pd.read_csv(r"C:\Users\buyt2\python\steps.csv")


print("Daily Steps Data:")
print(df)

# Basic analysis
print("\nAverage Steps:", df["Steps"].mean())
print("Max Steps in a Day:", df["Steps"].max())
print("Min Steps in a Day:", df["Steps"].min())

# Visualization
plt.plot(df["Day"], df["Steps"], marker='o', color='green')
plt.title("Daily Steps Over the Week")
plt.xlabel("Day")
plt.ylabel("Steps")
plt.grid(True)
plt.show()
```
